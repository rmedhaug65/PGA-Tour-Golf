```python
stats = list(golfer)
stats.remove('PlayerName')
stats.remove('Season')
```

```python
import matplotlib.pyplot as plt
#plt.rcParams['figure.figsize'] = [10, 8]

for column in stats:
    x = golfer['Season']
    y = golfer[column]
    fig, ax = plt.subplots()
    ax.plot(x, y, color='blue')
    ax.set(xlabel='Season', ylabel=column, title=player1 + "'s " + column)
    ax.grid()
    plt.show()
```
